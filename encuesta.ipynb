```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("./data.csv" , index_col=0)

data = data.sort_values("Muy interesado", ascending=False)
data = ((data * 100) / 2188).round(2)

Colors = ['#7CB518', '#F3DF2B','#FF1B1C' ]

ax = data.plot(kind='bar', 
               width=0.75, figsize=(20, 15), legend=True, fontsize=15, color=[Colors[0],Colors[1] ,Colors[2] ])
ax.set_title("Encuesta sobre el inters de las reas de las Ciencias de Datos", fontsize=20)
ax.set_ylabel('Porcentaje', fontsize=14)

for i in ax.patches:
    ax.text(i.get_x(), i.get_height()+ 2, \
            str(round(i.get_height(), 2))+'%', fontsize=15,
                color='black')

plt.show()
```
